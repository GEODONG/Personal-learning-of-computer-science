```python
#导入所需要的库
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Read excel
my_dataset = pd.read_excel('人工智能作业数据集.xlsx', sheet_name='Sheet5')
my_sub_dataset = my_dataset[['学习率','学习轮次','批次大小','准确率']]

#parameter name
lr = my_sub_dataset.学习率
num = my_sub_dataset.学习轮次
batch_size = my_sub_dataset.批次大小
Accuracy = my_sub_dataset.准确率

#plot figure
fig = plt.figure()
ax1 = plt.axes(projection='3d')
for i in range(0,32):
        if lr[i]==0.0001:
            my_color = 'black'
            z=Accuracy[i]
            y=num[i]
            x=batch_size[i]
            ax1.scatter3D(x,y,z,color = my_color,alpha=0.5)
        elif lr[i]==0.001:
            my_color = 'green'
            z=Accuracy[i]
            y=num[i]
            x=batch_size[i]
            ax1.scatter3D(x,y,z,color = my_color,alpha=0.5)
        elif lr[i]==0.01:
            my_color = 'yellow'
            z=Accuracy[i]
            y=num[i]
            x=batch_size[i]
            ax1.scatter3D(x,y,z,color = my_color,alpha=0.5)
        elif lr[i]==0.1:
            my_color = 'red'
            z=Accuracy[i]
            y=num[i]
            x=batch_size[i]
            ax1.scatter3D(x,y,z,color = my_color,alpha=0.5)
        elif lr[i]==1:
            my_color = 'blue'
            z=Accuracy[i]
            y=num[i]
            x=batch_size[i]
            ax1.scatter3D(x,y,z,color = my_color,alpha=0.5)

#set labels
ax1.set_xlabel('batch_size')
ax1.set_ylabel('num')
ax1.set_zlabel('Accuracy') 

plt.show()
```
